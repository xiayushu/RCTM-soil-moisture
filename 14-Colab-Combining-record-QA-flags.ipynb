# Combine quality control files from subsets deposited to the Google Cloud Bucket

```python
# Import quality control files for Modis GPP
# Customize the pathway based on where data is stored here and for following sections
!gsutil cp gs://bucket/Moisture/Test3/failgppset_*.csv .
```

```python
file_list = []
for files in glob.glob("failgppset*.csv"):
  file_list.append(files)
```

```python
gppresult_list = []
for file in file_list:
  try: 
    result = pd.read_csv(file, header = 0)
    siteresult = pd.DataFrame(result)
    gppresult_list.append(siteresult)
    gppresult = pd.DataFrame(np.concatenate(gppresult_list))
  except ValueError:
    continue
```

```python
gppresult = gppresult.iloc[:,[1]]
```

```python
gppresult.to_csv('1_final_gpp_result.csv') 
```
